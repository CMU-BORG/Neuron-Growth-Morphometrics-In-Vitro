```python
# This generates the violin plots found in Figure 7 in
# A S Liao, W Cui, V Webster-Wood, Y J Zhang
# Quantitative evaluation of neuron developmental morphology in vitro using the change-point test
# Submitted to: Neuroinformatics 2022
```

```python
#import relevant libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#read in the worksheet with the data to generate the violin plots
perCellMetrics = pd.read_csv('perCellMetrics(den10,rem8,remMiss).csv')

#image format options
image_format_svg = 'svg'
image_format_png = 'png'
figver = '_v03.'

saveFig = 1; #0 to not save, 1 to save
```

```python
f = plt.figure(figsize=(16,16)) #initialize figure
gs = f.add_gridspec(2,2) #2x2 subplot

sns.set(font_scale = 1.75, style = "ticks") #set seaborn figure features

#subplot in (0,0) is a violin plot the total length
ax1 = f.add_subplot(gs[0,0])
sns.violinplot(y="div",x="totLen",data=perCellMetrics,orient="h",inner="quart",cut=0)
sns.despine()
ax1.set_ylabel("Days in Vitro (DIV)")
ax1.set_xlabel("Total Length Per Cell (m)")
plt.xlim(0,1000)

#subplot in (0,1) is a violin plot of average segment length
ax2 = f.add_subplot(gs[0,1])
sns.violinplot(y="div",x="avgSegLen",data=perCellMetrics,orient="h",inner="quart",cut=0)
sns.despine()
ax2.set_ylabel("Days in Vitro (DIV)")
ax2.set_xlabel("Average Segment Length Per Cell (m)")
plt.xlim(0,25)

#subplot in (1,0) is a violin plot of degree
ax3 = f.add_subplot(gs[1,0])
sns.violinplot(y="div",x="degree",data=perCellMetrics,inner="quart",orient="h",cut=0)
sns.despine()
ax3.set_ylabel("Days in Vitro (DIV)")
ax3.set_xlabel("Degree Per Cell")
plt.xlim(0,16)
plt.xticks([0,4,8,12,16])

#subplot in (1,1) is a violin plot of the total number of change points
ax4 = f.add_subplot(gs[1,1])
sns.violinplot(y="div",x="totCPs",data=perCellMetrics,orient="h",inner="quart",cut=0)
sns.despine()
ax4.set_ylabel("Days in Vitro (DIV)")
ax4.set_xlabel("Number of Change Points Per Cell")
plt.xlim(0,120)

saveFigCombo = 0 #do not save this figure to the current directory if this is 0, save if 1
image_name = 'combo4'

if saveFigCombo == 1: #if saving the figure
    f.savefig(image_name+figver+image_format_svg,format=image_format_svg,dpi=1200) #save as svg
    f.savefig(image_name+figver+image_format_png,format=image_format_png,dpi=1200) #save as png
    print('Figure Saved!')
else:
    print('Figure Not Saved.')
```

```python
sns.set_style("ticks") #seaborn figure style

f,ax = plt.subplots(figsize=(8,8)) #figure set up

sns.violinplot(y="div",x="degree",data=perCellMetrics,inner="quart",orient="h",cut=0) #violin plot of degree

sns.despine(left=True) #remove left axis

#axes labeling
ax.set_ylabel("Days in Vitro (DIV)",size=16,alpha=0.7)
ax.set_xlabel("Degree Per Cell",size=16,alpha=0.7)
plt.xlim(0,16)
plt.xticks([0,4,8,12,16])

image_name = 'degree'

#save figure
if saveFig == 1:
    f.savefig(image_name+figver+image_format_svg,format=image_format_svg,dpi=1200)
    f.savefig(image_name+figver+image_format_png,format=image_format_png,dpi=1200)
    print('Figure Saved!')
else:
    print('Figure Not Saved.')
    
```
